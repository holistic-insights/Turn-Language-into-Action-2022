```python
import pandas as pd
import numpy as np
import requests
from googlesearch import search
import time
```

```python
list_startups = pd.read_csv("Companies.csv")
list_sp_500 = pd.read_csv("constituents_csv.csv")
list_euronext = pd.read_csv("Euronext_Equities.csv",sep=';')

#dropping irrelevant data
list_startups = list_startups.drop(["permalink","status","state_code","region","funding_total_usd"],axis=1)
list_startups = list_startups.iloc[:,0:6]
list_euronext = list_euronext.iloc[:,0].to_frame()
list_sp_500 = list_sp_500.drop(["Symbol"],axis=1)

list_startups = list_startups.rename(columns = {"name":"Name","market":"Sector","country_code":"HQ location"})

list_startups["Company stage"]= "Seed/Startup"
list_sp_500["Company stage"]= "ScaleUp/Mature"
list_euronext["Company stage"]= "ScaleUp/Mature"

list_sp_500["HQ location"]= "USA"
list_euronext["HQ location"]= "EU"
```

```python
#merging two databases 
df_companies = list_sp_500.append(list_startups)
df_companies = df_companies.append(list_euronext)
df_companies.head()
```

```python
#grabbing a sample of 250 companies
df_companies_summary_250 = df_companies.sample(n = 250)
df_companies_summary_50 = df_companies.sample(n = 50)
df_companies_summary_1000 = df_companies.sample(n = 1000)

df_companies.to_csv("df_companies_total.csv")
df_companies_summary_250.to_csv("df_companies_summary_250.csv")
df_companies_summary_50.to_csv("df_companies_summary_50.csv")
df_companies_summary_1000.to_csv("df_companies_summary_1000.csv")
```

```python
# creating function for website scraping of company list

def getURL(companyName, State):
    try:
        term = ' '.join([companyName, State])
        for url in search(term, num_results=1):
            return url
    except:
        return ''
```

```python
df_companies_summary_50['URL'] = ''

for index, row in df_companies_summary_50.iterrows():
    df_companies_summary_50.at[index, 'URL'] = getURL(row['Name'], row['HQ location'])
    time.sleep(2)
```
